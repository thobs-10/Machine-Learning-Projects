```python
from bs4 import BeautifulSoup
#use beautiful sooup to parse the html web page content
doc= BeautifulSoup(page_contents, 'html.parser')
```

```python
type(doc)
```

```python
# now get some information using queries
# lets grap the P tags from the web page
p_tags = doc.find_all('p')
len(p_tags)
```

```python
# now lets be more specific on which p tags we want because we grabbed to many
p_tags[:5]
```

```python
selected_class = "f3 lh-condensed mb-0 mt-1 Link--primary"

specific_tags = doc.find_all('p', {'class': selected_class})
```

```python
len(specific_tags)
```

```python
specific_tags
```

```python
# lets grap the topic ddescriptions
topic_title_descr_class = "f5 color-text-secondary mb-0 mt-1"
topic_title_descr_tags = doc.find_all('p', {'class': topic_title_descr_class})
```

```python
len(topic_title_descr_tags)
```

```python
topic_title_descr_tags[:5]
```

```python
# we need to get the topic url
topic_tag_link = doc.find_all('a', {'class': 'd-flex no-underline'})
len(topic_tag_link)
```

```python
# to get the href of the topic link 
topic_href_0 = topic_tag_link[0]['href']
topic_href_0
# now ltes try to combine github link with this link and see if it works
print('https://github.com'+topic_href_0)
```

```python
#get the topic title of every topic tag
topic_title =[]
for tag in specific_tags:
    topic_title.append(tag.text)
print(topic_title)
```

```python
# get topic descriptions in a list
topic_descriptions = []
for tag in topic_title_descr_tags:
    topic_descriptions.append(tag.text.strip())
topic_descriptions[:5]
```

```python
# grap topic url tags
topic_url = []
base_url = 'https://github.com'
for tag in topic_tag_link:
    topic_url.append(base_url + tag['href'])
topic_url
```

```python
# create a dataframe using pandas by taking this information into a dataframe
dict = {'topic' : topic_title,
       'description' : topic_descriptions,
       'url' : topic_url}

import pandas as pd
df = pd.DataFrame(dict)
df.head()
```

```python
doc_2 = BeautifulSoup(response_2.text, 'html.parser')
# find the repository name and user name and the number of stars and the link
selected_repo_tags = 'f3 color-text-secondary text-normal lh-condensed'
repo_tags = doc_2.find_all('h3', {'class': selected_repo_tags})
len(repo_tags)
```

```python
repo_name = a_tags[0].text.strip()
repo_name
```

```python
username = a_tags[1].text.strip()
username
```

```python
repo_link = a_tags[1]['href']
repo_link
```

```python
base_url = 'https://github.com'
print(base_url + repo_link)
```

```python
selected_star_tags = 'social-count float-none'
star_tags = doc_2.find_all('a', {'class': selected_star_tags})
len(star_tags)
```

```python
star = star_tags[0].text.strip()
star
```

```python
def parse_link_to_int(star_tag):
    star_tag.strip()
    if star_tag[-1] == 'k':
        #convert the value from string to int keep in mind the K=1000
        return int(float(star_tag[:-1]) * 1000)
    else:
        # if the star count is not thousand
        return int(star_tag)
```

```python
count = parse_link_to_int(star)
count
```

```python
# define a function that will get all the required information
def get_all_info(repository_tags, star_tag):
    #find all a tags 
    a_tags = repository_tags.find_all('a')
    #get the repo name
    repo_name = a_tags[0].text.strip()
    #get the username
    username = a_tags[1].text.strip()
    #get the star number
    star_count = parse_link_to_int(star_tag.text.strip())
    # get the repo url
    repo_url = base_url + a_tags[1]['href']
    return repo_name, username, star_count, repo_url
```

```python
get_all_info(repo_tags[0],star_tags[0])
```

```python
# now lets get all the information in a dictionary so we can pass it to a dataframe
repo_title_dict = {
    'repo_name':[],
    'username': [],
    'stars': [],
    'repo_url':[]
}

for i in range(len(repo_tags)):
    repo_info = get_all_info(repo_tags[i],star_tags[i])
    repo_title_dict['repo_name'].append(repo_info[0])
    repo_title_dict['username'].append(repo_info[1])
    repo_title_dict['stars'].append(repo_info[2])
    repo_title_dict['repo_url'].append(repo_info[3])
repo_title_dict
```
